**union**

```python
a = {1,2,3,4,5}
b = {4,5,6,7,8}

s = a.union(b) # union is use to combine the two set and removing the duplication value
s
```

**intersection**

```python
a = {1,2,3,4,5}
b = {4,5,6,7,8}

i = a.intersection(b) # intersection is used to combine same values in sets
i
```

**difference**

```python
a = {1,2,3,4,5}
b = {4,5,6,7,8}
da = a.difference(b)# these remove common values in one set and print reminding value
db = b.difference(a)
print(da)
print(db)
```

**symmetric_difference**

```python
a = {1,2,3,4,5}
b = {4,5,6,7,8}
s = a.symmetric_difference(b)# these remove common values in both set and print reminding value
s
```

## subset- is every itam of set a in set set b

**issuperset - set a and set b**

```python
a = {1,2,3,4}
b = {1,2,3,4,5,6}
a.issuperset(b)
```

```python
b.issuperset(a)
```
